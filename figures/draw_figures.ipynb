```python
import torch
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
loss_fully_datapath = '../exp_files/slide_loss_fully'
with open(loss_fully_datapath, 'rb') as f1:
    loss_fully = pickle.load(f1)
```

```python
loss_fully_dict = {'slide_loss': loss_fully, 'step': list(range(1, len(loss_fully)+1))}
```

```python
fig1, ax1 = plt.subplots(dpi=350)
ax1 = sns.lineplot(data=loss_fully_dict, x='step', y='slide_loss')
```

```python
fig1.savefig('fully_slide_loss.jpg')
```

```python
loss_hybrid_datapath = '../exp_files/slide_loss_hybrid.pkl'
with open(loss_hybrid_datapath, 'rb') as f1:
    loss_hybrid = pickle.load(f1)
loss_hybrid_dict = {'slide_loss': loss_hybrid, 'step': list(range(1, len(loss_hybrid)+1))}
```

```python
fig2, ax2 = plt.subplots(dpi=350)
ax2 = sns.lineplot(data=loss_hybrid_dict, x='step', y='slide_loss')
fig2.savefig('hybrid_slide_loss.jpg')
```
